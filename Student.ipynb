```python
import pandas as pd
import numpy as np
```

```python
df_cse=pd.read_csv("/content/CSE.csv")
df_it=pd.read_csv("/content/IT.csv")
df_student=pd.read_csv("/content/STUDENT.CSV")
df_cse.rename(columns={"Professional Elective":"PE"},inplace=True)
```

1.	Combine the CSE & IT data and display the data

```python
df_combine=pd.merge(df_cse,df_it,how="outer")
df_combine
```

2.	Display all CSE students’ marks along with personal information

```python
df1=pd.merge(df_cse,df_student,on="Regd.No")
df1
```

3.	Print all students Regd.No, Name and professional elective.

```python
df2=pd.merge(df_combine,df_student,on="Regd.No")
df2[['Regd.No','Name','PE']]
```

4.	Identify the students whose DAA marks are >18.

```python
df2[df2['DAA']>18][['Regd.No','Name']]
```

5.	Display the names and EAMCET ranks of the students who got minimum 12 marks in all courses.

```python
df2[(df2['DAA']>=12) &(df2['OE']>=12)&(df2['PE']>=12)&(df2['CN']>=12)&(df2['AFL']>=12)]['Name']
```

6.	Calculate mean value of all the subject’s marks.

```python
df2['AVG']=df2[['DAA','OE','PE','CN','AFL']].mean(axis=1)
df2['AVG']
```

7.	Display the names common in both CSE & IT along with Regd.No.

```python
dfcse=df2[df2['Branch']=='CSE']
dfit=df2[df2['Branch']=='IT']
dfn=pd.merge(dfcse,dfit)
dfn[['Name','Regd.No']]
```

8.	Fill the missing values of the data with  average marks of the subject of specific group.

```python
values={'DAA':np.mean(df2['DAA']),'AFL':np.mean(df2['AFL']),'CN':np.mean(df2['CN']),'OE':np.mean(df2['OE']),'PE':np.mean(df2['PE'])}
df2.fillna(value=values,inplace=True)
df2
```

9.	Divide the students into 5 groups based on average marks.

```python
gname=['Class I','Class II','Class III','Class IV','Class V']
df2['Category']=pd.cut(df2['AVG'],5,labels=gname)
print(pd.cut(df2['AVG'],5,labels=gname))
print(df2[['Name','Category']])
```

10.	Create  equal sized groups of students based on EAMCET Rank.

```python
gname=['Class I','Class II','Class III','Class IV','Class V']
df2['EAMCET_CAT']=pd.cut(df2['EAMCET RANK'],5,labels=gname)
print(df2[['Regd.No','EAMCET_CAT']])
```

11.	Display the electives and the Regd.No of students who opted the elective along with thesubject name.

```python
df_combine[df_combine['PE'].notnull()][['Regd.No','PE']]
```

12.	Compare the performance of the students from various cities.

```python
df2.groupby('Address')[['DAA','OE','PE','CN','AFL']].mean()
```
